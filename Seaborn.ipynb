## Scatter Plots

The scatter plot is mainstay of statistical visualization.
It depicts the joint distribution of two variables using cloud points, where each point represents an observation in the datset. This depiction allows the eye to infer a substantial amount
of information wheather there is any meaningful relationship between them.


Although, scatter plot can be plotted for both numerical as well as categorical data, it's mostly used
 to depict relationship between numerical data.
 
 There are 2 ways to draw a scatterplot in seaborn
 1. relplot [Figure Level Function]
 2. scatterplot [Axes Level Function]

As we will see, these visualizations can be quite helpful because they use simple and easily-
 understood representations of data even on complex data by mapping up to three additional
 variables using the semantics of hue, size, and style.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

### Plot a scatter plot between total_bill and tip

```python
# plot here
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
```

The Hue Parameter

```python
# plot here
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',data=data)
```

The Style parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',kind='scatter',data=data)
```

The Size parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',kind='scatter',style='sex',size='size',data=data)
```

# using the Scatter plotm

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```

# Heatmaps

A heatmap is a graphic al representation of data where the individual values contain in a matrix are represented as colors.

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
table = flights.pivot_table(index='year',columns ='month',values='passengers',aggfunc='sum')
table
```

```python
sns.heatmap(table,cmap='coolwarm')
```

```python
sns.heatmap(flights.cov())
```

```python
sns.heatmap(table,cbar=False)
```

```python
sns.heatmap(table,cbar=False,linewidth=0.5,annot=True,fmt='d')
```

```python
sns.heatmap(table,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(table,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='summer')
```

```python
sns.heatmap(table,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='winter')
```

```python
sns.heatmap(table,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='coolwarm')
```

# Clustermap

```python
sns.clustermap(table)
```

```python
sns.clustermap(table,metric='correlation')
```

```python
sns.clustermap(table, z_score=0,annot=True,row_cluster=False,cmap='viridis',metric='correlation')
```

```python
sns.clustermap(table, z_score=0,annot=True,col_cluster=False,cmap='coolwarm',metric='correlation')
```

# Distplots

Displot are used to do univariate analysis on a continous/numerical data

1. Histogram

2. Kernel Density Estimation(kde)

3. Rugplot

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.isnull().sum()
```

```python
titanic.age.fillna(round(titanic.age.mean()),inplace=True)
```

```python
titanic
```

```python
sns.displot(titanic.age,bins=20)
```

```python
sns.displot(titanic.age,kde=False)
```

```python
sns.displot(titanic.age,kind='kde')
```

```python
sns.displot(titanic.age,kind='kde',rug=True)
```

```python
titanic
```

How to use Distplot

```python
sns.displot(titanic[titanic.survived == 0])
```

```python
sns.displot(titanic[titanic.survived == 1]['age'],kde=True)
sns.displot(titanic[titanic.survived == 0]['age'],kde=True)
```
